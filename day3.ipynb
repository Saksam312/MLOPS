```python
import matplotlib.pyplot as plt

# Current admission percentage
admission_percentage = 100

# Test cases (increase percentages)
test_cases = [0, 20, 30, 40]

# Store results
new_percentages = []

print("Test Case Results")
print("-----------------")

for t in test_cases:
    result = admission_percentage * (1 + t / 100)
    new_percentages.append(result)
    print(f"Increase = {t}%  -->  Admission Percentage = {result}%")

# Plot graph
plt.plot(test_cases, new_percentages, marker='o', color='blue')
plt.title("Student Admission Percentage Increase")
plt.xlabel("Increase (%)")
plt.ylabel("Admission Percentage")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

admission_percentage = 100
increase = [0, 20, 30, 40]

new_percentage = [admission_percentage * (1 + i / 100) for i in increase]

sns.histplot(new_percentage, bins=4, kde=True, color="green")

plt.title("Distribution of Student Admission Percentage")
plt.xlabel("Admission Percentage")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Current admission percentage
current = 100

# Increase percentages
increase = [0, 20, 30, 40]

# Calculate next semester admission percentage
result = []

for i in increase:
    new_percentage = current + (current * i / 100)
    result.append(new_percentage)

# Test cases
assert result[0] == 100
assert result[1] == 120
assert result[2] == 130
assert result[3] == 140

print("All test cases passed")
print("Admission percentage:", result)

# Graph
plt.plot(increase, result, marker='o', color='green')

plt.title("Student Admission Percentage Increase")
plt.xlabel("Increase in Next Semester (%)")
plt.ylabel("Admission Percentage")

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Employee salary data
employee_data = {
    "Employee": ["E1", "E2", "E3", "E4", "E5"],
    "Department": ["CSE", "ECE", "EEE", "CSE", "ECE"],
    "Salary": [50000, 45000, 40000, 60000, 55000]
}

# Student fee data
student_data = {
    "Student": ["S1", "S2", "S3", "S4", "S5"],
    "Department": ["CSE", "ECE", "EEE", "CSE", "EEE"],
    "Fees": [80000, 75000, 70000, 85000, 72000]
}

employee_df = pd.DataFrame(employee_data)
student_df = pd.DataFrame(student_data)

# Total employee expense by department
salary_expense = employee_df.groupby("Department")["Salary"].sum()

# Total student fees by department
fee_collection = student_df.groupby("Department")["Fees"].sum()

# Combine both
analysis = pd.DataFrame({
    "Employee Salary Expense": salary_expense,
    "Student Fee Collection": fee_collection
})

# Sort department wise
analysis = analysis.sort_index()

print("Department Wise Analysis")
print(analysis)

# Plot
departments = analysis.index

colors = {
    "CSE": "red",
    "ECE": "blue",
    "EEE": "yellow"
}

plt.figure(figsize=(8,5))

for dept in departments:
    plt.bar(
        dept,
        analysis.loc[dept, "Employee Salary Expense"],
        color=colors[dept],
        label=f"{dept} Salary Expense"
    )

plt.xlabel("Department")
plt.ylabel("Amount")
plt.title("Employee Salary Expense by Department")
plt.legend()
plt.show()


# Student fee plot
plt.figure(figsize=(8,5))

for dept in departments:
    plt.bar(
        dept,
        analysis.loc[dept, "Student Fee Collection"],
        color=colors[dept],
        label=f"{dept} Fees"
    )

plt.xlabel("Department")
plt.ylabel("Amount")
plt.title("Student Fee Collection by Department")
plt.legend()
plt.show()
```
